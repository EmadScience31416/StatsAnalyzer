- Tablas utilizadas para la construcción:
1. Tabla Buró: Abril2022 - Enero 2023
2. Tabla Desembolsos: Junio 2021 - Diciembre 2022
3. Tabla Saldos: Junio - Diciembre 2022
4. Tabla Margen Orginal: Enero - Diciembre 2022
5. Tabla Cantidad Tipo Productos y Cantidad de Productos: al 31.12.2022

7. Tabla IdPrestamo vs IdSolicitud: Permite identificar si IdSolicitud desembolsó o no.

<B>Variables de Cliente con Clustering:</B> 
- SALDO_TOTAL_SF: Saldo total en el Sistema Financiero (en quetzales), al corte del mes de análisis. 
> Variable puede tomar valores de 0 a infinito positivo. Se construirá con Tabla de Buró. Se considera todos los productos (sin filtro en “TipoActivo”). Podemos tener registros en Buró con Saldo 0 (por ejemplo, son los créditos pagados con TC, o TC sin deuda al momento de consulta al Buró); estos se mantienen con 0 para saber que consultamos su Buró pero no tienen saldo en SF.
- 	SALDO_PREST_FIDUCIARIO_SF: Saldo de Préstamos Fiduciarios (préstamo personal o consumo) en el Sistema Financiero (en quetzales), al corte del mes de análisis.
> Variable puede tomar valores de 0 a infinito positivo. Se construirá con Tabla de Buró. Podemos tener registros en Buró con Saldo 0 (por ejemplo, son los créditos pagados con TC, o TC sin deuda al momento de consulta al Buró); estos se mantienen con 0 para saber que consultamos su Buró pero no tienen saldo en SF. Se considera “TipoActivo” = "Prestamos".  Para diferenciar el Tipo de Préstamo, se utiliza el filtro según lo informado por Interconsumo: “TipoGarantia” igual a "Fiduciaria", "Otras garantias" o "Fiduciaria-Otras garantias"

- SALDO_PREST_PRENDARIO_SF: Saldo de Préstamos Prendarios (o vehiculares) en el Sistema Financiero (en quetzales), al corte del mes de análisis.
> Variable puede tomar valores de 0 a infinito positivo. Se construirá con Tabla de Buró. Podemos tener registros en Buró con Saldo 0 (por ejemplo, son los créditos pagados con TC, o TC sin deuda al momento de consulta al Buró); estos se mantienen con 0 para saber que consultamos su Buró pero no tienen saldo en SF. 
Se considera “TipoActivo” = "Prestamos".  Para diferenciar el Tipo de Préstamo, se utiliza el filtro según lo informado por Interconsumo: “TipoGarantia” Prendas (solo o con mix)
- 	SALDO_PREST_HIPOTECARIO_SF: Saldo de Préstamos Hipotecarios en el Sistema Financiero (en quetzales), al corte del mes de análisis.
> Variable puede tomar valores de 0 a infinito positivo. Se construirá con Tabla de Buró. Podemos tener registros en Buró con Saldo 0 (por ejemplo, son los créditos pagados con TC, o TC sin deuda al momento de consulta al Buró); estos se mantienen con 0 para saber que consultamos su Buró pero no tienen saldo en SF. 
Se considera “TipoActivo” = "Prestamos". Para diferenciar el Tipo de Préstamo, se utiliza el filtro según lo informado por Interconsumo: “TipoGarantia” Inmuebles (solo o con mix)
- CANT_CREDITOS_VIGENTES_SF:  Cantidad de créditos vigentes en el Sistema Financiero, al corte del mes de análisis.
> Variable puede tomar valores desde 1 en adelante, salvo no tenga créditos vigentes al corte y sea 0. Se construirá con Tabla de Buró. Se considera solo “TipoActivo” "Prestamos" (sin filtro adicional). Se cuentan los registros por cada cliente.
- MONTO_DESEMB_TOTAL_SF_ULT6M: Monto desembolsado total en el sistema financiero, en los últimos 6 meses.
> Variable puede tomar valores mayores a 0, salvo no haya desembolsado nada en los últimos 6 meses y sea 0. Se construirá con Tabla de Buró, al corte del mes de análisis. Se considera solo “TipoActivo" "Prestamos” (sin filtro adicional). Usando la foto más reciente del Buró, se sumariza el “CapitalOriginal” de los créditos desembolsados en los últimos 6 meses (usando “FechaConcedido”). Considerar que:
		i. Solo podremos sumar los créditos desembolsados que siguen vigentes al corte del análisis. 
		ii. No se realiza filtro de clientes con Saldos 0 porque los consideramos créditos vigentes (aunque pagados).

- SHARE_OF_WALLET:  SALDO_TOTAL_IC / SALDO_TOTAL_SF. Se tendrá al corte del mes de análisis. 
> Variable puede tomar valores de 0 a 1. Se construirá con Tabla de Buró.  Se considera todos los productos (sin filtro en “TipoActivo”). Considerar que:
		i. Los nulos e indeterminados se considerarán como 0 en Share of Wallet bajo la mirada de Vinculación. 

- SHARE_OF_WALLET_FIDUCIARIO:  SALDO_PREST_FIDUCIARIO_IC / SALDO_PREST_FIDUCIARIO_SF. Se tendrá al corte del mes de análisis. 
> Variable puede tomar valores de 0 a 1. Se construirá con Tabla de Buró.  Se considera solo los préstamos Fiduciarios. Considerar que:
		i. Los nulos e indeterminados se considerarán como 0 en Share of Wallet bajo la mirada de Vinculación.

- SALDO_PROM_IC_6MESES: Promedio de Saldo Mensual de Cliente en Interconsumo en los últimos 6 meses.
> Variable puede tomar valores de 0 a infinito positivo. Se construirá con Tabla de Saldos. Se tendrá al corte del mes de análisis. Se considera todos los productos en la sumarización de saldo. Considerar que en la Tabla de Saldos podrían haber Saldos 0 porque son créditos vigentes que han sido pagados con TC pero aún deben pasar 3-4 meses para su cierre.
Todos los IdCliente que no figuren en la tabla de saldos en alguno de los 6 meses, tendrán el valor 0 en esta variables.
- CANT_DESEMBOLSOS_IC_6MESES: Cantidad de desembolsos del Cliente en Interconsumo en los últimos 6 meses. 
> Variable puede tomar valores mayores a 0, salvo que no haya desembolsado en los últimos 6 meses, donde sería 0. Se construirá con Tabla de Desembolsos. Se tendrá al corte del mes de análisis. Se decidió tomar periodo de 6 meses según input de Interconsumo.
- MONTO_DESEM_PRESTAMOS_IC_6MESES: Se sumariza los montos desembolsados de Cliente en Interconsumo en los últimos 6 meses. 
> Variable puede tomar valores mayores a 0, salvo que no haya desembolsado en los últimos 6 meses, donde sería 0. Se construirá con Tabla de Desembolsos. Se tendrá al corte del mes de análisis. Se decidió tomar periodo de 6 meses según input de Interconsumo.
- CANT_TIPO_PRODUCTOS_ACTIVOS:  Tipo de productos activos que tiene el cliente en Interconsumo (Consumo, Motos o Personal), al corte del mes de análisis. 
> Variable puede tomar valores de 0, 1, 2 o 3. Proviene de tabla "cantidad de tipo productos y cantidad de productos" calculada por Interconsumo. Todos los IdCliente que no figuren en la tabla "cantidad de tipo productos y cantidad de productos"  tendrán el valor 0 en esta variable.
- CANT_PRODUCTOS_ACTIVOS: Número de Productos Activos que tiene el cliente en Interconsumo, al corte del mes de análisis. 
> Variable puede tomar valores del 1 al 10, salvo no tenga productos activos al corte y en dicho caso sería 0. Proviene de tabla "cantidad de tipo productos y cantidad de productos"  calculada por Interconsumo. Todos los IdCliente que no figuren en la tabla "cantidad de tipo productos y cantidad de productos" tendrán el valor 0 en esta variable. Productos considerados: 'INTERCONSUMO PERSONAL', 'INTERCONSUMO FINANCIADO', 'INTERCONSUMO PERSONAL NANOCREDITO', 'ARTICULO RECUPERADO', 'INTERCONSUMO DE CONTADO', 'INTERCONSUMO DIFERIMIENTO', 'CONVENIO INTERBANCO PERSONAL', 'CONVENIO PERSONAL', 'CONVENIO CONSUMO' y ‘CONVENIO INTERBANCO CONSUMO’.
- MARGEN_ORDINAL_PROM_6MESES: 	Promedio de últimes 6 meses de Margen ordinal del cliente, al corte del mes de análisis. 
> Variable puede tomar valores negativos, cero y positivos. Tabla margen tiene la Rentabilidad mensual a nivel de Margen Ordinario +/- Recuperaciones y Castigos (Margen puede tomar valores negativos, cero y positivos). Cifras en quetzales. Todos los IdCliente que no figuren en la tabla Margen en los últimos 6 meses tendrán el valor 0 en esta variable.
- CANTIDAD_ENTIDADES_CON_DEUDA: Cantidad de entidades distintas con los que el cliente tiene deuda, al corte del mes de análisis. 
> Variable puede tomar valores de 1 a más, salvo no tenga deuda al corte y sea 0. Se construirá con Tabla de Buró. Se consideran todos los productos (sin filtro en “TipoActivo”). 
- 	CUPO_TC_SF: Limite total en Quetzales de tarjetas de crédito del cliente en el Sistema Financiero, al corte del mes de análisis. 
> Variable puede tomar valores mayores a 0, salvo no tenga TC al corte y sea 0. Se puede construir con Tabla de Buró. “CapitalOriginal" es Cupo o Límite de TC cuando “TipoActivo” es "Tarjeta de Credito". Se considera solo las líneas de “Moneda” “Quetzales”. Se considerará solo el “TipoActivo” "Tarjeta de Credito". Se suma el “CapitalOriginal" por cada cliente. Recordar que un cliente puede tener Tarjeta de crédito vigente con Saldo 0 o positivo en Buró.
- SALDO_PREST_FIDUC_COMPETIDORES: Saldo en préstamos fiduciarios en competidores directos (Banrural, Banco Azteca y Banco Antigua) al corte.
> Variable puede tomar valores de 0 a infinito positivo. Se construirá con Tabla de Buró. Podemos tener registros en Buró con Saldo 0 (por ejemplo, son los créditos pagados con TC, o TC sin deuda al momento de consulta al Buró); estos se mantienen con 0 para saber que consultamos su Buró pero no tienen saldo en SF. Se considera “TipoActivo” = "Prestamos".  Para diferenciar el Tipo de Préstamo, se utiliza el filtro según lo informado por Interconsumo: “TipoGarantia” igual a "Fiduciaria", "Otras garantias" o "Fiduciaria-Otras garantias".
- FLAG_USUARIO_DIGITAL: Si el Cliente tiene o no usuario en la aplicación móvil, al corte del mes de análisis. 
> Variable puede tomar valores de 0 (no tiene) o 1 (sí tiene). Proviene de Tabla Clientes. Como "TieneUsuarioDigital" viene a febrero 2023, para todos los registros donde el IdCliente tiene 'Alta Usuario' > '2022-12-31' imputaríamos 0 en la columna “FLAG_USUARIO_DIGITAL”.
- EDAD_ANIOS: Edad del Cliente en años, al corte del mes de análisis.
> Se construirá con Tabla Clientes. Se recalculará utilizando “FechaNacimiento” y Fecha de procesamiento para tener la edad al corte del mes de análisis. 
- ANTIGUEDAD_IC_CORTE: Antigüedad en días del cliente en Interconsumo, al corte del mes de análisis.
> Se tiene la columna "Antigüedad" con la foto de última solicitud en la Tabla Clientes. Se actualiza esta variable proyectando "Alta Solicitud" y fecha de procesamiento. En caso "IdPrestamo" para el "IdSolicitud" esté en NaN y "EstadoCliente" sea "POTENCIAL", se mantiene "ANTIGUEDAD_IC_CORTE" como 0.
- INGRESOS_QUETZALES: Ingresos mensuales del cliente en quetzales. Se tendrá solo la foto de última solicitud del cliente. 
> Variable puede tomar valores igual o mayor que Q1,000.00, e igual o menores que Q50,000.00. Proviene de Tabla Clientes.
- TIEMPO_LABORANDO_MESES: Tiempo que viene laborando el cliente (sea empleo formal o no). Se tendrá solo la foto de última solicitud del cliente. 
> Se filtrará registros con Nulos, negativos y mayor o igual a 612 meses (51 años), pues no hay validación del campo en sistema. Proviene de Tabla Clientes.

```python
# Para cada periodo y cliente calculamos los desembolsos aperturados en los últimos 6 meses 
lista_reporte = []
for fecha_reporte in df_buro_hist_prestamos.fechaSaldos.unique():
    df_buro_reporte = df_buro_hist_prestamos[df_buro_hist_prestamos.fechaSaldos == fecha_reporte]
    # Filtramos desembolsos aperturados en los últimos 6 meses   a aprtir de la fechaSaldos -> fechaSaldos_6m_antes <= FechaConcedido <=fechaSaldos_6m_antes
    cnd_6m_lower = (df_buro_reporte.FechaConcedido >=  df_buro_reporte.fechaSaldos_6m_antes)
    cnd_6m_upper = (df_buro_reporte.FechaConcedido <=  df_buro_reporte.fechaSaldos)
    df_buro_reporte = df_buro_reporte[cnd_6m_lower & cnd_6m_upper]
 
    
#     # Descomentar para validar con una fecha
#     print(df_buro_reporte.fechaSaldos.unique())
#     print(df_buro_reporte.FechaConcedido.unique())
#     display(df_buro_reporte.head())
#     break
#     Final de bloqeu de validacion
    
    # Eliminamos duplicados 
    df_buro_reporte.sort_values(["IdCliente", "FechaConcedido",  "fechaSaldos", "CapitalOriginal"], inplace = True)
    df_buro_reporte.drop_duplicates(["IdCliente", "FechaConcedido",  "fechaSaldos", "CapitalOriginal"], inplace = True, keep='first')
    
    # Sumarizamos el monto desembolsado por cada cliente 
    
    monto_desemb_cliente = df_buro_reporte[["IdCliente", "CapitalOriginal", "fechaSaldos"]].groupby(["IdCliente"], as_index=False).agg(
                   MONTO_DESEMB_TOTAL_SF_ULT6M = ('CapitalOriginal', 'sum'))
  
    monto_desemb_cliente['fechaSaldos'] = fecha_reporte
    lista_reporte.append(monto_desemb_cliente)

    
monto_desemb_cliente_mes =  pd.concat(lista_reporte)  
# Verificamos el cálculo
print('Filas:', monto_desemb_cliente_mes.shape[0], 'Columnas:', monto_desemb_cliente_mes.shape[1])
# Vemos las primeras filas
monto_desemb_cliente_mes.head()
```

```python
# Para cada periodo y cliente calculamos la suma de saldos de los últimos 6 meses 
lista_reporte_margen = []
for fecha_reporte in df_margen.Periodo_Fecha.unique():
    
    fecha_6m_antes = df_margen[df_margen.Periodo_Fecha == fecha_reporte]['Fecha_6m_antes'].unique()[0]
    
    # Filtramos el margen en los últimos 6 meses   a aprtir de Periodo_Fecha
    cnd_6m_lower = (df_margen.Periodo_Fecha >=  fecha_6m_antes)
    cnd_6m_upper = (df_margen.Periodo_Fecha <=  fecha_reporte)
    
    df_margen_reporte = df_margen[cnd_6m_lower & cnd_6m_upper]
    
    
  
    # Sumarizamos el margen por cada cliente 
    # Obtenemos las  MARGEN_ORDINAL_PROM_6MESES,CANT_ORDINAL_PROM_6MESES
    margen_cliente = df_margen_reporte.groupby(["Cod_cliente"], as_index=False).agg(
            CANT_ORDINAL_PROM_6MESES = ('Margen Ordinario', 'count'),
            MARGEN_ORDINAL_SUM_6MESES = ('Margen Ordinario', 'mean')  )
  
    margen_cliente['Periodo_Fecha'] = fecha_reporte
    lista_reporte_margen.append(margen_cliente)
    
margen_cliente_mes =  pd.concat(lista_reporte_margen) 
margen_cliente_mes['MARGEN_ORDINAL_PROM_6MESES'] = margen_cliente_mes['MARGEN_ORDINAL_SUM_6MESES']/periods
# Verificamos el cálculo
print('Filas:', margen_cliente_mes.shape[0], 'Columnas:', margen_cliente_mes.shape[1])
# Vemos las primeras filas
margen_cliente_mes
```
